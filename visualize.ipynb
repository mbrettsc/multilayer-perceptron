```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('data.csv')
df.tail()
```

```python
le = LabelEncoder()
df['M'] = le.fit_transform(df['M'])
df.tail()
```

```python
df.corr().head()
```

```python
sns.heatmap(df.corr());
```

```python
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(df.columns):
    axes[i].hist(df[df['M'] == 1][col], bins=30, alpha=0.3, label='M', color='red')
    axes[i].hist(df[df['M'] == 0][col], bins=30, alpha=0.5, label='B', color='blue')
    axes[i].set_title(col)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
target = df.iloc[:, 1].values
features = df.iloc[:, 2:].values 
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
pca_2d = PCA(n_components=2)
pca_2d_result = pca_2d.fit_transform(features_scaled)

pca_3d = PCA(n_components=3)
pca_3d_result = pca_3d.fit_transform(features_scaled)

tsne_2d = TSNE(n_components=2, random_state=42)
tsne_2d_result = tsne_2d.fit_transform(features_scaled)

tsne_3d = TSNE(n_components=3, random_state=42)
tsne_3d_result = tsne_3d.fit_transform(features_scaled)
```

```python
def plot_2d(data, title):
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=target, palette="coolwarm", s=60)
    plt.title(title)
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.legend(title="Class", labels=['Benign', 'Malignant'])
    plt.show()

def plot_3d(data, title):
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=target, cmap="coolwarm", s=60)
    ax.set_title(title)
    ax.set_xlabel("Component 1")
    ax.set_ylabel("Component 2")
    ax.set_zlabel("Component 3")
    legend1 = ax.legend(*scatter.legend_elements(), title="Class")
    ax.add_artist(legend1)
    plt.show()
```

```python
plot_2d(pca_2d_result, "PCA - 2D Projection")
```

```python
plot_2d(tsne_2d_result, "t-SNE - 2D Projection")
```

```python
plot_3d(pca_3d_result, "PCA - 3D Projection")
```

```python
plot_3d(tsne_3d_result, "t-SNE - 3D Projection")
```
